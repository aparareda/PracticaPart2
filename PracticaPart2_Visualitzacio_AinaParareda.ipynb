**3. Transformació i agregació de dades**

Per tal de preparar el conjunt de dades per a una visualització significativa i interactiva, en aquest pas duré a terme diverses transformacions que permeten obtenir nous indicadors a partir de les dades originals. Aquestes transformacions són clau per extreure coneixement útil i visualitzar patrons o desigualtats que no serien evidents en els valors bruts.

**Creació de noves variables**
- Percentatge d'homes i dones per ensenyament: Aquesta variable permet detectar els estudis més feminitzats o masculinitzats.

```python
# Percentatge d'homes respecte el total d’alumnes per fila
df['percentatge_homes'] = df['Matrícules. Homes'] / df['Matrícules. Total'] * 100
# Percentatge de dones respecte el total d’alumnes per fila
df['percentatge_dones'] = df['Matrícules. Dones'] / df['Matrícules. Total'] * 100
```

- Alumnes per unitat (densitat educativa): Aquest indicador mesura la “pressió educativa” d’un ensenyament, és a dir, quants alumnes hi ha per cada grup/unitat ofert. Pot permetre detectar possibles desequilibris en la ràtio alumnes/grups. Serà molt útil per identificar territoris amb saturació.

```python
# Densitat educativa: alumnes per unitat (assumint que 1 unitat és el mínim raonable)
df['alumnes_per_unitat'] = df.apply(
    lambda row: row['Matrícules. Total'] / (row['Unitats'] if row['Unitats'] > 0 else 1),
    axis=1
)
```

- Índex de segregació de gènere: Una manera simple d’estimar com de polaritzat està un ensenyament en termes de gènere. S’apropa a 0 si hi ha equilibri, i a 1 si hi ha domini absolut d’un sol gènere. Ens permet identificar quins estudis presenten més biaix de gènere de forma objectiva i comparable.

```python
# Absolut de la diferència percentual entre homes i dones, dividit per 100
df['index_segregacio_genere'] = abs(df['percentatge_dones'] - df['percentatge_homes']) / 100
```

**4. Visualització exploratòria**

En aquest apartat realitzaré agregacions de dades per facilitar la posterior visualització i l'anàlisi de patrons territorials i educatius. Faré agrupacions per municipi, comarca i tipus d’ensenyament per poder observar desigualtats o diferències entre territoris o nivells educatius.
Pretenc obtenir indicadors agregats com la mitjana d’alumnes per unitat o el percentatge de dones.
Amb aquests càlculs permetran detectar patrons territorials i estructurals dins del sistema educatiu.

```python
# 4. Agrupació i càlcul d'indicadors
import pandas as pd
from IPython.display import display

# Agrupació per municipi: mitjana d’alumnes per unitat
print("\nMitjana d’alumnes per unitat per municipi (Top 10):")
alumnes_per_unitat_municipi = df.groupby('Nom municipi')['alumnes_per_unitat'].mean().reset_index()
alumnes_per_unitat_municipi = alumnes_per_unitat_municipi.sort_values(by='alumnes_per_unitat', ascending=False)
display(alumnes_per_unitat_municipi.head(10))

# Agrupació per comarca: percentatge mitjà de dones
print("\nPercentatge mitjà de dones per comarca (Top 10):")
percentatge_dones_comarca = df.groupby('Nom comarca')['percentatge_dones'].mean().reset_index()
percentatge_dones_comarca = percentatge_dones_comarca.sort_values(by='percentatge_dones', ascending=False)
display(percentatge_dones_comarca.head(10))

# Agrupació per comarca: percentatge mitjà d'homes'
print("\nPercentatge mitjà d'homes per comarca (Top 10):")
percentatge_homes_comarca = df.groupby('Nom comarca')['percentatge_homes'].mean().reset_index()
percentatge_homes_comarca = percentatge_homes_comarca.sort_values(by='percentatge_homes', ascending=False)
display(percentatge_homes_comarca.head(10))

# Agrupació per nivell agrupat: mitjana d’alumnes per unitat
print("\nMitjana d’alumnes per unitat segons els estudis:")
mitjana_alumnes_nivell = df.groupby('Codi estudis')['alumnes_per_unitat'].mean().reset_index()
mitjana_alumnes_nivell = mitjana_alumnes_nivell.sort_values(by='alumnes_per_unitat', ascending=False)
display(mitjana_alumnes_nivell)

# Agrupació per ensenyament: índex de segregació de gènere mitjà
print("\nÍndex mitjà de segregació de gènere per ensenyament (Top 10 amb més segregació):")
df['index_segregacio'] = abs(df['percentatge_dones'] - 50) / 50
segregacio_per_ensenyament = df.groupby('Nom ensenyament')['index_segregacio'].mean().reset_index()
segregacio_per_ensenyament = segregacio_per_ensenyament.sort_values(by='index_segregacio', ascending=False)
display(segregacio_per_ensenyament.head(10))
```

**5. Narrativa visual**

L’objectiu d’aquesta anàlisi és comprendre millor la realitat del sistema educatiu a Catalunya mitjançant dades obertes. Ens preguntem:

- **Hi ha segregació de gènere en els estudis?**

Per a analitzar-ho creo un gràfic de barres que mostra el percentatge de nois i noies per cada ensenyament ordenat per segregació. És rellevant ja que la segregació de gènere en estudis pot reproduir-se al món laboral i perpetuar desigualtats estructurals.

Amb els estudis que hem fet anteriorment hem vist que hi ha modalitats amb només un o dos alumnes (sobretot disciplines d'algun instrument musical poc comú) que mostren segregació per gènere, en aquest cas filtrem perquè mínim tingui 10 alumnes.

```python
import plotly.graph_objects as go

# Agrupació i càlculs
segregacio = df.groupby('Nom ensenyament').agg({
    'Matrícules. Homes': 'sum',
    'Matrícules. Dones': 'sum'
}).reset_index()

# Total i filtre
segregacio['Total'] = segregacio['Matrícules. Homes'] + segregacio['Matrícules. Dones']
segregacio = segregacio[segregacio['Total'] > 10]  # només estudis amb pes real

# Percentatges
segregacio['% Homes'] = (segregacio['Matrícules. Homes'] / segregacio['Total']) * 100
segregacio['% Dones'] = (segregacio['Matrícules. Dones'] / segregacio['Total']) * 100

# Índex de segregació
segregacio['Index segregació'] = abs(segregacio['% Homes'] - segregacio['% Dones'])

# Ordeno pel grau de desequilibri
segregacio = segregacio.sort_values('Index segregació', ascending=False).head(30)

# Gràfic apilat horitzontal
fig = go.Figure()

fig.add_trace(go.Bar(
    y=segregacio['Nom ensenyament'],
    x=segregacio['% Homes'],
    name='% Homes',
    orientation='h',
    marker_color='steelblue',
    hovertemplate='% Homes: %{x:.1f}%<extra></extra>'
))

fig.add_trace(go.Bar(
    y=segregacio['Nom ensenyament'],
    x=segregacio['% Dones'],
    name='% Dones',
    orientation='h',
    marker_color='lightpink',
    hovertemplate='% Dones: %{x:.1f}%<extra></extra>'
))

fig.update_layout(
    barmode='stack',
    title="Distribució de gènere per ensenyament (ordenat per més segregació)",
    xaxis_title='Percentatge',
    yaxis_title='Nom de l\'ensenyament',
    legend_title='Gènere',
    height=1000
)

fig.show()

```

Observem que els estudis tècnics (FP industrial, informàtica) són majoritàriament masculins, mentre que estudis socials o sanitaris són majoritàriament femenins. Això indica un biaix de gènere persistent.

- **Hi ha pressió educativa en determinats territoris o centres?**

Amb aquest gràfic podem veure en quines comarques la mitjana d'alumnes per unitat és més alta. Això pot voler dir que hi ha més demanda educativa i que les classes, per cada unitat, són més nombroses, estan més plenes.

```python
import plotly.express as px

# Converteixo els noms de comarca a majúscula per evitar duplicats per format
df['Nom comarca'] = df['Nom comarca'].str.upper()

df_comarques = df.groupby('Nom comarca')['alumnes_per_unitat'].mean().reset_index()

fig2 = px.bar(
    df_comarques.sort_values('alumnes_per_unitat', ascending=False).head(20),
    x='alumnes_per_unitat',
    y='Nom comarca',
    orientation='h',
    color='alumnes_per_unitat',
    color_continuous_scale='YlOrRd',
    labels={'alumnes_per_unitat': 'Alumnes per unitat'},
    title='Comarques amb més densitat educativa'
)
fig2.show()
```

En aquest cas veiem que la majoria de les ràtios amb més alumnes es troba en l'àrea metropolitana, amb l'excepció del Segrià.

Per a analitzar si té a veure amb l'oferta de diferents tipus d'estudis realitzo un mapa de calor segons el tipus d'estudis i les comarques.

```python
# Agrupo les dades
heatmap_data = df.groupby(['Nom comarca', 'Codi estudis'])['Matrícules. Total'].sum().reset_index()

# Converteixo a majúscules per evitar duplicats
heatmap_data['Nom comarca'] = heatmap_data['Nom comarca'].str.upper()
pivot_df = heatmap_data.pivot(index='Codi estudis', columns='Nom comarca', values='Matrícules. Total').fillna(0)

# Gràfica amb les dimensions
fig5 = px.imshow(
    pivot_df,
    labels=dict(x="Comarca", y="Tipus d'estudis (Codi)", color="Nombre d'alumnes"),
    color_continuous_scale='YlGnBu',
    title="Mapa de calor: distribució dels estudis per comarca"
)
fig5.update_layout(xaxis={'categoryorder': 'total descending'})
fig5.show()
```

Observem que, com era d'esperar, les unitats amb més alumnes són les d'educacions obligatòries, com l'educació primària o secundària. I que on hi ha més varietat en les educacions és, també, en l'àrea metropolitana.



```python
# Normalitzo el nom de les comarques i titularitat
df['Nom comarca'] = df['Nom comarca'].str.upper()
df['Nom titularitat'] = df['Nom titularitat'].str.upper()

# Agrupo per comarca
comarques_agg = df.groupby('Nom comarca')['Matrícules. Total'].sum().reset_index()
comarques_ordenades = comarques_agg.sort_values('Matrícules. Total')

comarca_min = comarques_ordenades.iloc[0]['Nom comarca']
comarca_max = comarques_ordenades.iloc[-1]['Nom comarca']
comarca_mitjana = comarques_ordenades.iloc[len(comarques_ordenades) // 2]['Nom comarca']

comarques_objetiu = [comarca_min, comarca_mitjana, comarca_max]

for comarca in comarques_objetiu:
    print(f"\nGràfiques per la comarca: {comarca.title()}")

    # TOTAL
    sub_total = df[df['Nom comarca'] == comarca].groupby('Codi estudis')['Matrícules. Total'].sum().reset_index()
    sub_total = sub_total.sort_values('Matrícules. Total', ascending=False).head(15)

    fig_total = px.pie(
        sub_total,
        values='Matrícules. Total',
        names='Codi estudis',
        title=f"[Total] Matrícules per codi d’estudis a {comarca.title()}",
        color_discrete_sequence=px.colors.sequential.Blues
    )
    fig_total.show()

    # PÚBLIC
    sub_public = df[
        (df['Nom comarca'] == comarca) & (df['Nom naturalesa'].str.contains("Públic"))
    ].groupby('Codi estudis')['Matrícules. Total'].sum().reset_index()
    sub_public = sub_public.sort_values('Matrícules. Total', ascending=False).head(15)

    fig_public = px.pie(
        sub_public,
        values='Matrícules. Total',
        names='Codi estudis',
        title=f"[Pública] Matrícules per codi d’estudis a {comarca.title()}",
        color_discrete_sequence=px.colors.sequential.Greens
    )
    fig_public.show()

    # PRIVAT
    sub_privat = df[
        (df['Nom comarca'] == comarca) & (df['Nom naturalesa'].str.contains("Privat"))
    ].groupby('Codi estudis')['Matrícules. Total'].sum().reset_index()
    sub_privat = sub_privat.sort_values('Matrícules. Total', ascending=False).head(15)

    fig_privat = px.pie(
        sub_privat,
        values='Matrícules. Total',
        names='Codi estudis',
        title=f"[Privada] Matrícules per codi d’estudis a {comarca.title()}",
        color_discrete_sequence=px.colors.sequential.OrRd
    )
    fig_privat.show()

```

- **Hi ha segregació segons públic o privat?**

Analitzo si hi ha diferències notables entre les inscripcions en centres públics o privats.


```python
import plotly.express as px

df_naturalesa_genere = df.groupby('Nom naturalesa')[['Matrícules. Dones', 'Matrícules. Homes']].sum().reset_index()

fig1 = px.bar(
    df_naturalesa_genere.melt(id_vars='Nom naturalesa', var_name='Gènere', value_name='Matrícules'),
    x='Nom naturalesa',
    y='Matrícules',
    color='Gènere',
    barmode='group',
    title='Distribució de matrícules per gènere segons naturalesa del centre (Públic/Privat)',
    labels={'Nom naturalesa': 'Naturalesa del centre'},
    color_discrete_map={
        'Matrícules. Dones': 'hotpink',
        'Matrícules. Homes': 'dodgerblue'
    }
)
fig1.show()
```

No s'observen diferències notables de segregació de gènere segons públic o privat.

Analitzem, també, si per tipus d'estudis sí que n'hi ha:

```python
# Agrupo les matrícules per Codi d'estudis i naturalesa (Públic/Privat)
df_estudis_naturalesa = df.groupby(['Codi estudis', 'Nom naturalesa'])[['Matrícules. Total']].sum().reset_index()

import plotly.express as px

fig = px.bar(
    df_estudis_naturalesa,
    x='Codi estudis',
    y='Matrícules. Total',
    color='Nom naturalesa',
    barmode='group',
    title='Distribució de matrícules per Codi d’estudis i naturalesa del centre (Públic/Privat)',
    color_discrete_map={
        'Públic': 'green',
        'Privat': 'orange'
    },
    labels={'Codi estudis': 'Codi d’estudis', 'Nom naturalesa': 'Naturalesa del centre'}
)

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```
